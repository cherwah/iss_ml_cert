```python
stop_words = stopwords.words('english') 

lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(docs):
    docs_clean = []
    punc = str.maketrans('', '', string.punctuation)
    for doc in docs:
        doc_no_punc = doc.translate(punc)
        words = doc_no_punc.lower().split()    
        words = [lemmatizer.lemmatize(word, 'v')
                        for word in words if word not in stop_words]    
        docs_clean.append(' '.join(words))
    
    return docs_clean

docs = preprocess(corpus)
```

```python
tfidf = TfidfVectorizer()
tfidf.fit(docs)
tfidf_docs = tfidf.transform(docs).toarray()
```

```python
features = tfidf.get_feature_names()
indexes = ['doc'+str(i) for i in range(len(corpus))]

tfidf_df = pd.DataFrame(data=tfidf_docs, index=indexes, columns=features)
tfidf_df
```

```python
query = ['life wise choices']
query_arr = preprocess(query)

query_arr
```

```python
tfidf_query = tfidf.transform(query_arr).toarray()
tfidf_query

tfidf_q_df = pd.DataFrame(data=tfidf_query, index=['query'], columns=features)
tfidf_q_df
```

```python
tfidf_q_df.loc[:,['life', 'wise', 'choices']]
```

```python
docs_similarity = cosine_similarity(tfidf_q_df, tfidf_df)
docs_similarity

cs = pd.DataFrame(data=docs_similarity,
            columns=['doc'+str(i+1) for i in range(60)])

cs
```

```python
query_similarity = docs_similarity[0]
query_similarity
```

```python
series = pd.Series(query_similarity, index=tfidf_df.index)
sorted_series = series.sort_values(ascending=False)
sorted_series = sorted_series[sorted_series!=0]

sorted_series
```

```python
print("\nSearch results for query: '", query, "':\n", sep='')

for index in sorted_series.index:
	doc_idx = int(index[3:])
	print(corpus[doc_idx], " [score = ", sorted_series[index], "]\n", sep='')
```
